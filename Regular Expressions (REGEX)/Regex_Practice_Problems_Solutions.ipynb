# 1.Write a Python program to check that a string contains only a certain set of characters (in this case a-z, A-Z and 0-9)

```python
def check_alphanumeric(input_string):
    """
    Checks if a string contains only alphanumeric characters (a-z, A-Z, 0-9).
    """
    # Use a regular expression to match strings containing only a-z, A-Z, and 0-9
    # The re.fullmatch() function in Python's re module attempts to match the entire string with the regular expression pattern. This means that for fullmatch() to return a match, the entire input string must match the pattern. If the pattern matches only a part of the string, or if there are additional characters before or after the match, fullmatch() will return None.
    # ^ asserts position at the start of the string
    # [a-zA-Z0-9] matches any character from the set
    # * matches the previous token zero or more times
    # $ asserts position at the end of the string
    if re.fullmatch(r'^[a-zA-Z0-9]*$', input_string):
        return True
    else:
        return False

# Example usage:
print(f"'HelloWorld123' contains only alphanumeric characters: {check_alphanumeric('HelloWorld123')}")
print(f"'Hello World!' contains only alphanumeric characters: {check_alphanumeric('Hello World!')}")
print(f"'Python_3' contains only alphanumeric characters: {check_alphanumeric('Python_3')}")
print(f"'' contains only alphanumeric characters: {check_alphanumeric('')}")
```

# 2.Write a Python program that matches a string that has an a followed by zero or more b's

```python
def check_a_followed_by_b(input_string):
    """
    Checks if a string has an 'a' followed by zero or more 'b's.
    """
    # The regex 'ab*' matches 'a' followed by zero or more 'b's.
    # re.fullmatch ensures the entire string matches the pattern.
    if re.fullmatch(r'ab*', input_string):
        return True
    else:
        return False

# Example usage:
print(f"'a' matches 'ab*': {check_a_followed_by_b('a')}")
print(f"'ab' matches 'ab*': {check_a_followed_by_b('ab')}")
print(f"'abbb' matches 'ab*': {check_a_followed_by_b('abbb')}")
print(f"'ac' matches 'ab*': {check_a_followed_by_b('ac')}")
print(f"'ba' matches 'ab*': {check_a_followed_by_b('ba')}")
print(f"'' matches 'ab*': {check_a_followed_by_b('')}")
```

# 7. Write a Python program to find sequences of lowercase letters joined by an underscore

```python
import re

def find_lowercase_sequences_with_underscore(input_string):
    """
    Finds all sequences of one or more lowercase letters joined by an underscore.
    """
    # The regex '[a-z]+_[a-z]+' matches one or more lowercase letters
    # followed by an underscore, followed by one or more lowercase letters.
    # re.findall returns all non-overlapping matches of the pattern in the string.
    matches = re.findall(r'[a-z]+_[a-z]+', input_string)
    return matches

# Example usage:
print(f"In 'hello_world', sequences are: {find_lowercase_sequences_with_underscore('hello_world')}")
print(f"In 'first_name_last_name', sequences are: {find_lowercase_sequences_with_underscore('first_name_last_name')}")
print(f"In 'noUnderscoreHere', sequences are: {find_lowercase_sequences_with_underscore('noUnderscoreHere')}")
print(f"In 'UPPER_CASE_UNDERSCORE', sequences are: {find_lowercase_sequences_with_underscore('UPPER_CASE_UNDERSCORE')}")
print(f"In 'single_char_a_b_c', sequences are: {find_lowercase_sequences_with_underscore('single_char_a_b_c')}")
print(f"In 'number_123_test', sequences are: {find_lowercase_sequences_with_underscore('number_123_test')}")
```

# 8. Write a Python program to find the sequences of one upper case letter followed by lower case letters

```python
import re

def find_uppercase_followed_by_lowercase(input_string):
    """
    Finds all sequences of one uppercase letter followed by one or more lowercase letters.
    """
    # [A-Z] matches one uppercase letter
    # [a-z]+ matches one or more lowercase letters
    matches = re.findall(r'[A-Z][a-z]+', input_string)
    return matches

# Example usage:
print(f"In 'Python Exercises', sequences are: {find_uppercase_followed_by_lowercase('Python Exercises')}")
print(f"In 'AaBbCc', sequences are: {find_uppercase_followed_by_lowercase('AaBbCc')}")
print(f"In 'apple', sequences are: {find_uppercase_followed_by_lowercase('apple')}")
print(f"In 'USA', sequences are: {find_uppercase_followed_by_lowercase('USA')}")
print(f"In 'Hello world, this is a Test', sequences are: {find_uppercase_followed_by_lowercase('Hello world, this is a Test')}")
```

# 10. Write a Python program that matches a word at the beginning of a string

```python
import re

def match_word_at_beginning(input_string):
    """
    Matches a word at the beginning of a string.
    """
    # ^ matches the start of the string
    # \w+ matches one or more alphanumeric characters or underscores
    match = re.match(r'^\w+', input_string)
    if match:
        return match.group()
    else:
        return None

# Example usage:
print(f"Word at beginning of 'Hello World': {match_word_at_beginning('Hello World')}")
print(f"Word at beginning of 'Python_Programming': {match_word_at_beginning('Python_Programming')}")
print(f"Word at beginning of '  Leading Space': {match_word_at_beginning('  Leading Space')}")
print(f"Word at beginning of '!Special Character': {match_word_at_beginning('!Special Character')}")
```

### 11. Write a Python program that matches a word at the end of a string, with optional punctuation

```python
import re

def match_word_at_end(input_string):
    """
    Matches a word at the end of a string, including optional punctuation.
    """
    # \w+ matches word characters
    # [.,!?;:]* matches optional punctuation characters at the end
    # $ ensures it is at the very end of the string
    match = re.search(r'\w+[.,!?;:]*$', input_string)
    if match:
        return match.group()
    return None

# Example usage:
print(f"End word of 'The exercise is fun.': {match_word_at_end('The exercise is fun.')}")
print(f"End word of 'Are you ready?': {match_word_at_end('Are you ready?')}")
print(f"End word of 'Look at that!': {match_word_at_end('Look at that!')}")
print(f"End word of 'No punctuation': {match_word_at_end('No punctuation')}")
```

# 12. Write a Python program that matches a word containing 'z'.

```python
import re

def find_words_containing_z(input_string):
    """
    Finds all words in a string that contain the letter 'z'.
    """
    # \b matches a word boundary
    # \w* matches zero or more word characters
    # z matches the literal 'z'
    # re.IGNORECASE is used to match both 'z' and 'Z'
    matches = re.findall(r'\b\w*z\w*\b', input_string, re.IGNORECASE)
    return matches

# Example usage:
text = "The lazy brown fox jumps over the amazed zoo keepers."
print(f"Words containing 'z' in '{text}':")
print(find_words_containing_z(text))

text2 = "Pizza is amazing, but zebras are cool too."
print(f"\nWords containing 'z' in '{text2}':")
print(find_words_containing_z(text2))
```

# 14. Write a Python program to match a string that contains only upper and lowercase letters, numbers, and underscores

```python
import re

def check_string_composition(input_string):
    """
    Checks if a string contains only letters, numbers, and underscores.
    """
    # ^ asserts start of string
    # \w+ matches one or more alphanumeric or underscore characters
    # $ asserts end of string
    if re.fullmatch(r'^\w+$', input_string):
        return True
    else:
        return False

# Example usage:
print(f"'Valid_String123' matches: {check_string_composition('Valid_String123')}")
print(f"'Invalid-String!' matches: {check_string_composition('Invalid-String!')}")
print(f"'Space Not Allowed' matches: {check_string_composition('Space Not Allowed')}")
print(f"'only_underscores' matches: {check_string_composition('only_underscores')}")
```

# 15. Write a Python program that starts each string with a specific number

```python
import re

def starts_with_specific_number(input_string, number):
    """
    Checks if the input string starts with the specified number.
    """
    # ^ specifies the start of the string
    # str(number) converts the input number to its string representation
    # r (Raw String): This prefix tells Python to treat backslashes as literal characters. While not strictly necessary for this specific number pattern, it is a best practice in regex to avoid issues with Python's escape sequences (like \n).
    # f (Formatted String): This allows you to embed variables directly into the string using curly braces {}. If number is 5, f'^{number}' becomes the string '^5'.
    pattern = rf'^{number}'
    if re.match(pattern, input_string):
        return True
    else:
        return False

# Example usage:
test_number = 5
print(f"'5-star rating' starts with {test_number}: {starts_with_specific_number('5-star rating', test_number)}")
print(f"'Stars: 5' starts with {test_number}: {starts_with_specific_number('Stars: 5', test_number)}")
print(f"'500 internal error' starts with {test_number}: {starts_with_specific_number('500 internal error', test_number)}")
```

# 16. Write a Python program to remove leading zeros from an IP address

```python
import re

def remove_leading_zeros_ip(ip_address):
    """
    Removes leading zeros from an IP address.
    Example: 192.168.001.009 -> 192.168.1.9
    """
    # \b matches a word boundary
    # 0+ matches one or more zeros
    # (\d) matches a digit and captures it in group 1
    # r'\1' replaces the match with the captured digit
    return re.sub(r'\b0+(\d)', r'\1', ip_address)

# Example usage:
ip1 = "192.168.001.009"
ip2 = "100.020.003.400"
ip3 = "010.000.000.010"

print(f"Original: {ip1} -> Cleaned: {remove_leading_zeros_ip(ip1)}")
print(f"Original: {ip2} -> Cleaned: {remove_leading_zeros_ip(ip2)}")
print(f"Original: {ip3} -> Cleaned: {remove_leading_zeros_ip(ip3)}")
```

# 17. Write a Python program to check for a number at the end of a string.

```python
import re

def check_number_at_end(input_string):
    """
    Checks if a string ends with a digit (0-9).
    """
    # \d matches any digit (0-9)
    # $ asserts position at the end of the string
    if re.search(r'\d$', input_string):
        return True
    else:
        return False

# Example usage:
print(f"'Exercise 1' ends with a number: {check_number_at_end('Exercise 1')}")
print(f"'Exercise' ends with a number: {check_number_at_end('Exercise')}")
print(f"'Python 3.10' ends with a number: {check_number_at_end('Python 3.10')}")
print(f"'Number123!' ends with a number: {check_number_at_end('Number123!')}")
```

# 18. Write a Python program to search for numbers (0-9) of length between 1 and 3 in a given string

```python
import re

def find_numbers_1_to_3_digits(input_string):
    """
    Searches for numbers (0-9) of length between 1 and 3 in a string.
    """
    # \b ensures we match the whole number, not part of a larger one
    # \d{1,3} matches 1, 2, or 3 digits
    return re.findall(r'\b\d{1,3}\b', input_string)

# Example usage:
text = "Exercises 1, 12, 123, and 1234 are present here, along with 56."
print(f"Numbers with 1-3 digits in: '{text}'")
print(find_numbers_1_to_3_digits(text))
```

# 19. Write a Python program to search for literal strings within a string.

Sample text : 'The quick brown fox jumps over the lazy dog.'
Searched words : 'fox', 'dog', 'horse'

```python
import re

def search_literals(text, searched_words):
    """
    Searches for a list of literal strings within a text.
    """
    for word in searched_words:
        print(f"Searching for '{word}' in text...")
        if re.search(word, text):
            print(f"-> Match found!")
        else:
            print(f"-> No match found.")

# Example usage:
sample_text = 'The quick brown fox jumps over the lazy dog.'
searched_words = ['fox', 'dog', 'horse']

search_literals(sample_text, searched_words)
```

# 20. Write a Python program to search for a literal string and find its location in the text

```python
import re

def find_literal_location(text, word):
    """
    Searches for a word and returns its start and end indices.
    """
    match = re.search(word, text)
    if match:
        start = match.start()
        end = match.end()
        return f"Found '{word}' at index {start} to {end}."
    else:
        return f"'{word}' not found."

# Example usage:
sample_text = 'The quick brown fox jumps over the lazy dog.'
print(find_literal_location(sample_text, 'fox'))
print(find_literal_location(sample_text, 'dog'))
print(find_literal_location(sample_text, 'horse'))
```
