```python
# Importing library 

import re
```

```python
text_to_search = '''

abcdefghijklmnopqrstuvwxyz
ABCDEFGHIJKLMNOPQRSTUVWXYZ
1234567890

Ha HaHa

MetaCharacters (Need to be escaped):

. ^ $ * + ? { } [ ] \ | ( )

coreyms.com

321-555-4321
123.555.1234
123*555*1234

Mr. Schafer
Mr Smith
Ms Davis
Mrs. Robinson
Mr. T

'''

sentence = 'Start a sentence and then bring it to an end'
```

- d and D, w and W, s and S: Capital letters negate the small characters output
- "-" is a special character. when it is put at the beginning or in [] character set it will just match the literal - character
    but when placed between the values it can specify a range of values*

```python
# . - Any charcter except new line

'''
returns the iterable object because we are using iter function

span gives the indexes of the pattern match

match shows the pattern we opted to find


"." represents every single character from the search string and returns the index of the every character except new line. 
for reference check the below: 28,29 index will not be shown because its the new/next line

<re.Match object; span=(27, 28), match='z'>
<re.Match object; span=(29, 30), match='A'>
'''

pattern = re.compile(r'.')

matches = pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# \d  -  matches all the digits (0-9)
'''
\d matches all the numbers between 0 and 9. below is the output gives the span (indexes) of the number present in the search 
string
'''

pattern = re.compile(r'\d')

matches = pattern.finditer (text_to_search)

for match in matches:
    print(match)
```

```python
# \D  -  not a digit (0-9)
'''
\D does not match any of the numbers 0 and 9. below is the output gives the span (indexes) except numbers present in the search 
string
'''

pattern = re.compile(r'\D')

matches = pattern.finditer (text_to_search)

for match in matches:
    print(match)
```

```python
# \w  -  word character (a-z, A-Z,0-9,_)
'''
\w matches word character (a-z, A-Z,0-9,_). below is the output gives the span (indexes) present excluding special
characters in the search string
'''

pattern = re.compile(r'\w')

matches = pattern.finditer (text_to_search)

for match in matches:
    print(match)
```

```python
# \W  -  not a word character (a-z, A-Z,0-9,_)
'''
\W does not matches word character (a-z, A-Z,0-9,_). below is the output gives the span (indexes) present only the special
characters in the search string
'''

pattern = re.compile(r'\W')

matches = pattern.finditer (text_to_search)

for match in matches:
    print(match)
```

```python
# \s  -  white space (space, tab, newline)
'''
\s matches white space (space, tab, newline). below is the output gives the span (indexes)
'''

pattern = re.compile(r'\s')

matches = pattern.finditer (text_to_search)

for match in matches:
    print(match)
```

```python
# \S  - not white space (space, tab, newline)
'''
\S does not matches white space (space, tab, newline). below is the output gives the span (indexes)
'''

pattern = re.compile(r'\S')

matches = pattern.finditer (text_to_search)

for match in matches:
    print(match)
```

# Anchors :

They don't actually match any characters but rather invisible positions before or after the characters. We can use this in conjunction with other patterns for searching

```python
# \b -  word boundary 

'''
\b word boundary. Word boundary are indicated by whitespace  or non alphanumeric character 

below is the output gives the span (indexes) to search for "Ha"

output only matches - Ha Ha(partial word) because of boundary
'''

pattern = re.compile(r'\bHa')

matches = pattern.finditer (text_to_search)

for match in matches:
    print(match)
```

```python
# \B -  not a word boundary 

'''
\B not a word boundary. 

below is the output gives the span (indexes) to search for "Ha"

output only matches - Ha from Ha(Ha) because it does last 2 letters does not have any boundary. Partial data will be matched
'''

pattern = re.compile(r'\BHa')

matches = pattern.finditer (text_to_search)

for match in matches:
    print(match)
```

```python
# ^ -  beginning of the string

'''
^ check the beginning of the string 

below is the output gives the span (indexes) to search for start. 

when searched for "a" it does give any output because string starts with S not with a in Sentence variable

'''

pattern = re.compile(r'^Start')

matches = pattern.finditer (sentence)

for match in matches:
    print(match)
    
    
pattern = re.compile(r'^a')

matches = pattern.finditer (sentence)

for match in matches:
    print(match)    
```

```python
# $ -  end of a string

'''
^ check the end of the string 

below is the output gives the span (indexes) to search for end. 

when searched for "a" it does give any output because string ends with end not with a in Sentence variable

'''

pattern = re.compile(r'end$')

matches = pattern.finditer (sentence)

for match in matches:
    print(match)
    
    
pattern = re.compile(r'a$')

matches = pattern.finditer (sentence)

for match in matches:
    print(match)   
```

# Quantifiers

- To match more than 1 character at once
- Ex: while search for phone number we are passing \d multiple times to match the phone format. Instead we can use quantifiers 
    to match the data and same output
    
- " * " -- matches 0 or more
- "+" -- 1 or more
- ? -- 0 or 1
- {3} --  exact 3 
- {3,4} --  Range of numbers (Minimum, Maximum)

```python
# get the number using quantifiers

pattern = re.compile(r'\d{3}.\d{3}.\d{4}')
matches =  pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

# Group Patterns

Groups allows us to match several different patterns so to create a group we use parenthesis

- "|" -- Either or 
-  () -- Group 

```python
# fetch all the names which starts with Mr/Ms/Mrs

pattern = re.compile(r'M(r|s|rs)\.?\s[A-Z]\w*') # we created Mr, Ms, Mrs as 3 seperate groups for searching
matches =  pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# fetch all the names which starts with Mr/Ms/Mrs

pattern = re.compile(r'(Mr|Ms|Mrs)\.?\s[A-Z]\w*')

# we created Mr, Ms, Mrs as 3 seperate groups for searching. Instead of Putting M outside we kept inside for readability

matches =  pattern.finditer(text_to_search)

for match in matches:
    print(match)
```

```python
# fetch the domains from below urls

urls ='''
https://www.google.com
http://coreyms.com
https://youtube.com
https://www.nasa.gov
'''


pattern = re.compile(r'https?://(www\.)?\w+\.\w+')

matches =  pattern.finditer(urls)

for match in matches:
    print(match)





```

```python
# split the below repression in to 3 different groups such as www. | websitename | domain. there won't be any change in o/p

pattern = re.compile(r'https?://(www\.)?(\w+)(\.\w+)')

'''
(www\.)? - pull data even www. is available or not
(\w+) - website name
(\.\w+) - domain name -- .com|.gov

'''

matches =  pattern.finditer(urls)

for match in matches:
    print(match)

    
```

```python
# split the below repression in to 3 different groups such as www. | websitename | domain. there won't be any change in o/p

pattern = re.compile(r'https?://(www\.)?(\w+)(\.\w+)')

'''
(www\.)? - pull data even www. is available or not
(\w+) - website name
(\.\w+) - domain name -- .com|.gov

now in the match function we can pass group function which will pull data based on the groups created during compile

group(0)- return entire url
group (1) -  return www. or none ( if not present)
group (2) - return website name
group (3) - return domain name

'''

matches =  pattern.finditer(urls)

for match in matches:
    print(match.group(0))
    print(match.group(1))
    print(match.group(2))
    print(match.group(3))
```

-  We can use something called back reference to reference our captured group and it's basically just a shorthand for accessing
    these group indexes 
- So the regular expression module has a sub method that we can use to perform a substitution 

### Use Case of substitutions:

- if we have a large dataset of things we wanted to reformat like this then using regular expressions could save time and allow
    us to do in couple of minutes


```python
# fetch the domains from below urls

urls ='''
https://www.google.com
http://coreyms.com
https://youtube.com
https://www.nasa.gov
'''


pattern =  re.compile(r'https?://(www\.)?(\w+)(\.\w+)')

subbed_urls = pattern.sub(r'\2\3',urls) # we are substituting/replacing the 2 and 3 groups when the entire urls is fetched

print(subbed_urls)
```

### Findall Method:

- with the finditer method - it returns match objects with extra information and functionality but findall will just return the
    matches as a list of strings
- now if its matching groups then it will only return groups 


```python
'''
- in the below example we are currently using the pattern where we match those names and we have a group at the beginning for 
    the prefix. so this is only going to match that group
-  if we have multiple groups then it will return as list of tuples and the tuples will contain all of the groups
-  now if there are no groups then it would retun all of the matches in a list of strings
'''



pattern = re.compile(r'(Mr|Ms|Mrs)\.?\s[A-Z]\w*') # we created Mr, Ms, Mrs as 3 seperate groups for searching
matches =  pattern.findall(text_to_search)

for match in matches:
    print(match)
```

# Match Method:

- Match function will determine if the regular expression matches at the beginning of the string 
-  it will throw an error stating it wont return iterable object like finditer or findall
-  it just returns the first match and if there is no match it returns none
- so instead of using for loop we can direct using print command
- it will search only the beginnning of the string not the entire string

```python
pattern = re.compile(r'Start')

matches = pattern.match(sentence)

print(matches)
```

# Search Method:

- Match method will search only the begining of the string where as search function searches the entire string
-  it only prints the first occurence of the search string just like match function

```python
pattern = re.compile(r'sentence')

matches = pattern.search(sentence)

print(matches)
```

# Flags:

-  Let's say we wanted to match a word but match it whether  it was in upper case or lower case or mixture of both 

```python
# Ignorecase ignores the casing. We can use IGNORECASE or I 
pattern = re.compile(r'start', re.IGNORECASE)

matches = pattern.search(sentence)

print(matches)



pattern = re.compile(r'start', re.I)

matches = pattern.search(sentence)

print(matches)
```
