```python
# ==============================
# 1. Install & Import Libraries
# ==============================
# Run this once in terminal:
# pip install spacy emoji contractions
# python -m spacy download en_core_web_sm

import re
import string
import spacy
import emoji
import contractions

# Load spaCy model
nlp = spacy.load("en_core_web_sm")
```

```python
# ==============================
# 2. Sample Dataset
# ==============================
text = """
OMG!!! This product is sooo goooood 😍😍!!! 
I bought it from https://example.com on 12/12/2024.
Contact me at test@email.com or call 9876543210.
#Amazing #BestProduct Ever!!! I'm loving it!!!
"""
```

```python
# ==============================
# 3. Basic Cleaning Functions
# ==============================

def to_lowercase(text):
    return text.lower()

def remove_urls(text):
    return re.sub(r'http\S+|www\S+', '', text)

def remove_emails(text):
    return re.sub(r'\S+@\S+', '', text)

def remove_phone_numbers(text):
    return re.sub(r'\b\d{10}\b', '', text)

def remove_html_tags(text):
    return re.sub(r'<.*?>', '', text)

def remove_emojis(text):
    return emoji.replace_emoji(text, replace='')

def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))

def remove_numbers(text):
    return re.sub(r'\d+', '', text)

def remove_extra_whitespace(text):
    return re.sub(r'\s+', ' ', text).strip()
```

```python
# ==============================
# 4. Advanced Cleaning
# ==============================

def expand_contractions(text):
    return contractions.fix(text)

# Reduce repeated characters (soooo → soo)
def reduce_lengthening(text):
    return re.sub(r'(.)\1+', r'\1\1', text)

def remove_hashtags_mentions(text):
    text = re.sub(r'#\w+', '', text)
    text = re.sub(r'@\w+', '', text)
    return text
```

```python
# ==============================
# 5. spaCy Processing
# ==============================

def spacy_processing(text):
    doc = nlp(text)

    tokens = []
    lemmas = []
    pos_tags = []
    entities = []

    for token in doc:
        tokens.append(token.text)
        lemmas.append(token.lemma_)
        pos_tags.append((token.text, token.pos_))

    for ent in doc.ents:
        entities.append((ent.text, ent.label_))

    return tokens, lemmas, pos_tags, entities
```

```python
# ==============================
# 6. Stopword Removal
# ==============================

def remove_stopwords_spacy(tokens):
    return [token for token in tokens if not nlp.vocab[token].is_stop]
```

```python
# ==============================
# 7. Full Preprocessing Pipeline
# ==============================

def preprocess_text(text):
    print("Original Text:\n", text)

    # Basic cleaning
    text = to_lowercase(text)
    text = expand_contractions(text)
    text = remove_urls(text)
    text = remove_emails(text)
    text = remove_phone_numbers(text)
    text = remove_html_tags(text)
    text = remove_emojis(text)
    text = remove_hashtags_mentions(text)
    text = reduce_lengthening(text)
    text = remove_punctuation(text)
    text = remove_numbers(text)
    text = remove_extra_whitespace(text)

    print("\nCleaned Text:\n", text)

    # spaCy NLP processing
    tokens, lemmas, pos_tags, entities = spacy_processing(text)

    print("\nTokens:\n", tokens)
    print("\nLemmas:\n", lemmas)
    print("\nPOS Tags:\n", pos_tags)
    print("\nNamed Entities:\n", entities)

    # Stopword removal
    tokens_no_stop = remove_stopwords_spacy(tokens)
    print("\nAfter Stopword Removal:\n", tokens_no_stop)

    return {
        "clean_text": text,
        "tokens": tokens,
        "no_stopwords": tokens_no_stop,
        "lemmas": lemmas,
        "pos_tags": pos_tags,
        "entities": entities
    }
```

```python
# ==============================
# 8. Run Pipeline
# ==============================

output = preprocess_text(text)

print(output)
```
